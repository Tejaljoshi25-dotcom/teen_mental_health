## LITERATURE REVIEW:
1. These foundational insights are drawn from established, public academic health frameworks and peer-reviewed journals, mapping out the existing scientific baseline before exploring the current dataset:

2. The Dose-Response Effect of Screen Time: Global psychiatric and pediatric consensus points to a distinct, non-linear relationship between high screen engagement and compromised psychological well-being. According to clinical studies published by the Mayo Clinic, adolescents who cross the threshold of 3 hours of daily social platform engagement exhibit a significantly heightened risk for internalizing clinical problems, including anxiety and mood instability.

3. Gender-Specific Digital Discrepancy: Epidemiological surveys conducted by the World Health Organization (WHO) alongside independent longitudinal studies (e.g., PMC9954802) confirm a structural gender divide in digital behavioral patterns. Problematic social media usage, compulsive checking loops, and cognitive dependency show higher nodes among adolescent females, driven heavily by social comparison dynamics, whereas adolescent males tend to show greater clustering inside competitive digital gaming cycles.

4. The Melatonin Suppression & Sleep Loop: Clinical research by specialized pediatric institutions, including the Royal Children’s Hospital (RCH), confirms that evening exposure to short-wavelength blue light emitted from digital devices severely delays the synthesis of melatonin. The resulting compression of the sleep window compromises crucial overnight cognitive restoration, acting as a direct neurobiological catalyst that amplifies daytime stress and emotional dysregulation.

5. Age and Neurodevelopmental Dynamics: Because the prefrontal cortex—the region of the brain governing impulse control, long-term gratification tracking, and emotional regulation—is not fully matured during the teenage years, adolescents remain structurally vulnerable to tech-induced dependency loops. As individuals transition into mid-adolescence (15-17 years), the developmental imperative for peer validation peaks, often generating sharp increases in device and platform addiction levels.

## Plot 1: The Gender Digital Usage Paradox
Traditional studies kehti hain ki girls zyada screen time consume karti hain, par is data mein humne ek bilkul alag trend 
dekha hai. Iske liye humne standard bar chart ke bajaye ek sophisticated Violin Plot banaya hai jo pure density structure ko
breakdown karta hai.

```python
import warnings

# Sabhi future warnings ko completely suppress karein
warnings.simplefilter(action='ignore', category=FutureWarning)

# Plot customization aur background themes
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# 1. Left Panel: Daily Social Media Hours
sns.barplot(
    data=df, 
    x='gender', 
    y='daily_social_media_hours', 
    ax=axes[0], 
    palette=['#3498db', '#e74c3c'], 
    errorbar=None, 
    hue='gender'
)
axes[0].set_title('Average Daily Social Media Usage', fontsize=12, fontweight='bold', pad=10)
axes[0].set_xlabel('Gender', fontsize=11)
axes[0].set_ylabel('Hours per Day', fontsize=11)

# 2. Right Panel: Screen Time Before Sleep
sns.barplot(
    data=df, 
    x='gender', 
    y='screen_time_before_sleep', 
    ax=axes[1], 
    palette=['#3498db', '#e74c3c'], 
    errorbar=None, 
    hue='gender'
)
axes[1].set_title('Bedtime Screen Time Exposure', fontsize=12, fontweight='bold', pad=10)
axes[1].set_xlabel('Gender', fontsize=11)
axes[1].set_ylabel('Hours Before Sleep', fontsize=11)

# Overall clean title header
plt.suptitle('Plot 1: Gender-Wise Digital Footprint (At a Glance)', fontsize=15, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

INSIGHTS: No Outperformance Paradox: Dono graphs ke bar heights ko dekhte hi pata chalta hai ki digital patterns mein koi bada gender gap nahi hai.

The Symmetrical Loop: Ladke average routine check mein daily screen time par thode se high hain, par jaise hi raat ka sleep window aata hai, text/browse screen nodes par female indicators light marginal margin se heavy ho jate hain.

```python
## Plot 2: Age vs. Device Addiction Level.
# Calculate mean addiction level per age group
age_addiction = df.groupby('age')['addiction_level'].mean().reset_index()

# Plot setup
sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 5))

# Plot 2: Bar plot showing Age vs Average Addiction Level
ax = sns.barplot(
    data=age_addiction, 
    x='age', 
    y='addiction_level', 
    palette='Reds_d'
)

# Add exact value annotations on top of each bar for instant scannability
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(
            f"{p.get_height():.2f}", 
            (p.get_x() + p.get_width() / 2., p.get_height()), 
            ha='center', va='bottom', 
            fontsize=11, fontweight='bold', 
            xytext=(0, 3), textcoords='offset points'
        )

plt.title('Plot 2: Average Device Addiction Level Across Ages (13-19)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Age of Teenager', fontsize=12)
plt.ylabel('Average Addiction Rating (1-10)', fontsize=12)
plt.ylim(0, 7)  # Adjust Y-axis limit so data labels fit nicely
plt.tight_layout()
plt.show()
```

```python
INSIGHTS : The 13-Year-Old Initial Peak (5.75): Early teenagers exhibit a surprisingly high addiction rating (~5.75), which 
           directly reflects initial unrestricted smartphone access and curiosity-driven consumption.

The Mid-Teen Stability Window (14-15): Ages 14 and 15 drop slightly to ~5.20–5.28, showing relative stability in habit structures.

The Late Adolescent Surge (16-19): From age 16 onwards, device addiction steadily escalates, reaching its peak at Age 19 (5.91)
This aligns with academic literature where higher independence, university transitions, and constant peer networking drive
screen dependency.
```

```python
# Categorize daily social media hours into 3 distinct tiers
df['social_media_tier'] = pd.cut(
    df['daily_social_media_hours'], 
    bins=[0, 3, 6, 9], 
    labels=['1-3 Hrs (Low)', '3-6 Hrs (Medium)', '6-8 Hrs (High)']
)

# Calculate group means for Stress and Anxiety
tier_mental = df.groupby('social_media_tier', observed=False)[['stress_level', 'anxiety_level']].mean().reset_index()

# Reshape data for clean grouped bar plotting
df_melted = tier_mental.melt(
    id_vars=['social_media_tier'], 
    value_vars=['stress_level', 'anxiety_level'],
    var_name='Indicator', 
    value_name='Score'  # Fixed parameter name
)

df_melted['Indicator'] = df_melted['Indicator'].replace({
    'stress_level': 'Average Stress Level',
    'anxiety_level': 'Average Anxiety Level'
})

# Plot 3 setup
plt.figure(figsize=(10, 5.5))

ax = sns.barplot(
    data=df_melted, 
    x='social_media_tier', 
    y='Score', 
    hue='Indicator', 
    palette=['#e67e22', '#9b59b6']
)

# Add exact value labels on top of each bar
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(
            f"{p.get_height():.2f}", 
            (p.get_x() + p.get_width() / 2., p.get_height()), 
            ha='center', va='bottom', 
            fontsize=11, fontweight='bold', 
            xytext=(0, 3), textcoords='offset points'
        )

plt.title('Plot 3: Daily Social Media Consumption Tiers vs. Stress & Anxiety', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Daily Social Media Usage Tier', fontsize=12)
plt.ylabel('Average Score (1-10 Scale)', fontsize=12)
plt.ylim(0, 8)
plt.legend(title='Psychological Indicator')
plt.tight_layout()
plt.show()
```

 1-The Anxiety Threshold Spike (3–6 Hours Tier):

- Observation: Anxiety levels jump from 5.45 in the Low Usage tier to their peak at 5.80 in the Medium Usage tier (3–6 hours).

- Insight: Crossing the 3-hour daily threshold acts as an immediate psychological trigger, causing a noticeable spike in anxiety.

2- The High-Usage Stress Surge (6–8 Hours Tier):

- Observation: Stress levels escalate steadily from 5.20 in the Medium tier to 5.76 in the High Usage tier (6–8 hours).

- Insight: Chronic, heavy social media consumption (over 6 hours daily) directly correlates with the highest average stress burden among adolescents.

3- Inversion of Dominant Psychological Strain:

- Low Tier (1–3 Hrs): Stress (5.51) is slightly higher than Anxiety (5.45).

- Medium Tier (3–6 Hrs): Anxiety (5.80) significantly overtakes Stress (5.20).

- High Tier (6–8 Hrs): Stress (5.76) rebounds to become the dominant strain over Anxiety (5.58).

```python
plt.figure(figsize=(8, 5.5))

ax = sns.scatterplot(
    data=df, 
    x='sleep_hours', 
    y='stress_level', 
    hue='depression_label', 
    palette=['#3498db', '#e74c3c'], 
    alpha=0.8,
    s=70
)

plt.title('Plot 4: Sleep Duration vs. Stress Level (Target: Depression Status)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Sleep Hours', fontsize=12)
plt.ylabel('Stress Level (1-10 Scale)', fontsize=12)

# Fix legend labels for clear readability
handles, _ = ax.get_legend_handles_labels()
plt.legend(handles=handles, labels=['No Depression (0)', 'Depressed (1)'], title='Target Status')

plt.tight_layout()
plt.show()
```

1. The High-Risk Depression Quadrant:

- Observation: Every single teenager flagged positive for depression (1) is strictly concentrated in the top-left corner of the plot.

- Exact Metrics: Depressed teens average only 4.76 hours of sleep (Range: 4.1 to 5.8 hours) and an extreme average stress level of 8.48 / 10 (Range: 7 to 10).

- Insight: Depressive symptoms do not occur randomly across the population; they are tightly bound to severe chronic stress combined with acute sleep deprivation.

2. The 6-Hour Protective Sleep Threshold:

- Observation: There are zero red points beyond the 5.8-hour mark on the sleep axis.

- Insight: Getting 6 or more hours of nightly sleep acts as a clear physiological protective threshold, below which the probability of a positive depression flag drops to zero in this dataset.

3. High Stress Alone vs. Sleep Synergy:

- Observation: Many non-depressed teens (blue dots) also experience high stress levels (8–10), but they get 6 to 9 hours of sleep.

- Insight: High stress alone is not enough to trigger a depression flag—it is the synergistic combination of high stress AND low sleep (< 6 hrs) that creates the clinical vulnerability window.

```python
## Plot 6: Physical Activity Tiers vs. Stress & Anxiety Levels
# Categorize daily physical activity into 3 clear tiers
df['activity_tier'] = pd.cut(
    df['physical_activity'], 
    bins=[-0.1, 0.5, 1.2, 2.1], 
    labels=['0-0.5 Hrs (Low)', '0.5-1.2 Hrs (Moderate)', '1.2-2.0 Hrs (Active)']
)

# Aggregate mean scores
act_mental = df.groupby('activity_tier', observed=False)[['stress_level', 'anxiety_level']].mean().reset_index()

# Reshape data for grouped bar chart
df_melted = act_mental.melt(
    id_vars=['activity_tier'], 
    value_vars=['stress_level', 'anxiety_level'],
    var_name='Indicator', 
    value_name='Score'
)

df_melted['Indicator'] = df_melted['Indicator'].replace({
    'stress_level': 'Average Stress Level',
    'anxiety_level': 'Average Anxiety Level'
})

# Plot 6
plt.figure(figsize=(8, 4))

ax = sns.barplot(
    data=df_melted, 
    x='activity_tier', 
    y='Score', 
    hue='Indicator', 
    palette=['#2ecc71', '#34495e']
)

# Add value labels on top of bars
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(
            f"{p.get_height():.2f}", 
            (p.get_x() + p.get_width() / 2., p.get_height()), 
            ha='center', va='bottom', 
            fontsize=11, fontweight='bold', 
            xytext=(0, 3), textcoords='offset points'
        )

plt.title('Plot 6: Physical Activity Tiers vs. Stress & Anxiety Levels', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Daily Physical Activity Tier', fontsize=12)
plt.ylabel('Average Score (1-10 Scale)', fontsize=12)
plt.ylim(0, 8)
plt.legend(title='Psychological Indicator')
plt.tight_layout()
plt.show()
```

1- Anxiety Drop in Moderate Activity: Teens with low physical activity (0–0.5 hrs) exhibit the highest average anxiety score (5.85). Increasing physical activity to 0.5–1.2 hours drops anxiety to 5.50.

2- Stress Stability: Stress scores remain remarkably stable across all activity brackets (~5.25 – 5.55), confirming that physical activity primarily impacts anxiety levels rather than general perceived stress in this dataset

3. Literature Review & Theoretical Foundations
- The Dose-Response Effect: Clinical guidelines from the Mayo Clinic indicate that crossing 3 hours of daily social media engagement significantly elevates internalizing mental health problems, such as anxiety and mood dysregulation.

- Gender Parity in Consumption: Surveys by the World Health Organization (WHO) highlight that while digital engagement is high across all adolescents, dependency and usage loops differ structurally across platforms and mediums.

- Melatonin & Sleep Disruption: Pediatric research from the Royal Children’s Hospital (RCH) confirms that evening blue-light exposure delays melatonin secretion, compressing the sleep window and compromising overnight neurological restoration.

- Neurodevelopmental Vulnerability: Brain development studies confirm that the prefrontal cortex remains immature through late adolescence, leaving teens (15–19 years) susceptible to compulsive device checking and digital addiction.

5. Key Findings & Empirical Results
Finding 1: Gender Consumption Symmetry (Plot 1 & Plot 5)
- Social Media Hours: Males average 4.56 hours/day, while females average 4.51 hours/day.

- Bedtime Exposure: Females report an average of 1.76 hours before sleep vs. 1.72 hours for males.

- Conclusion: Digital consumption in adolescents is a universal generational habit rather than a gender-segregated behavior.

Finding 2: Age-Wise Addiction Dynamics (Plot 2)
- Early Peak: Age 13 exhibits an initial addiction score of 5.75 / 10 due to new device access.

- Mid-Teen Stabilization: Ages 14–15 drop to 5.20–5.28 / 10.

- Late Adolescent Escalation: Device addiction surges from age 16 onward, peaking at 5.91 / 10 at Age 19.

Finding 3: Psychological Strain & Usage Thresholds (Plot 3)
- The 3-Hour Anxiety Threshold: Anxiety scores jump from 5.45 (Low Tier: 1–3 hrs) to 5.80 (Medium Tier: 3–6 hrs) as soon as daily usage crosses 3 hours.

- Heavy Usage Stress Surge: Daily consumption exceeding 6 hours pushes average stress levels to their highest point (5.76 / 10).

  Finding 4: The Sleep-Deprivation Depression Cluster (Plot 4)

- High-Risk Zone: All positive depression flags (depression_label = 1, n=31) are exclusively clustered in the low sleep duration (4.1–5.8 hours) and severe stress (7–10 / 10) quadrant.

- Protective Boundary: Adolescents securing $\ge$ 6 hours of nightly sleep demonstrate zero depression flags in this dataset.

Finding 5: Physical Activity Modulation (Plot 6)
- Engaging in 0.5–1.2 hours of daily physical activity lowers baseline anxiety scores from 5.85 to 5.50 / 10.

  6. Conclusion & Actionable Recommendations
- Enforce the 3-Hour Screen Limit: Interventions should target keeping daily social media usage below 3 hours to prevent baseline anxiety spikes.

- Protect the 6-Hour Sleep Floor: Ensuring at least 6 hours of uninterrupted sleep serves as a vital physiological barrier against stress-induced depressive symptoms.

- Target Mid-to-Late Teens: Digital literacy and screen management campaigns should focus on high school seniors and university entrants (Ages 16–19).
